```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import seaborn as sns
#import tensorflow as tf
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split 
from sklearn.metrics import mean_squared_error as mse, r2_score as r2e, accuracy_score, confusion_matrix, classification_report
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset_cols = ["bike_count", "hour", "temp", "humidity", "wind", "visibility", "dew_pt_temp", "radiation", "rain", "snow","seasons", "holiday", "functioning_day"]
df = pd.read_csv("~/projects/prices-predictor-system/maverick-predictor-system/extracted_data/SeoulBikeData.csv").drop(['Date'], axis=1)
df.columns = dataset_cols
#df["functional"] = (df["functional"] == "Yes").astype(int)

df.head()
```

```python
for x_label in df.columns:
   plt.figure(figsize=(10,6))
   sns.histplot(df[x_label],bins=30, kde=True)
   plt.xlabel("Target value")
   plt.ylabel("frequency")
   plt.show()

   plt.figure(figsize=(12,8))
   sns.boxplot(x=df[x_label], y=df["bike_count"]) if x_label != "bike_count" else None
   plt.show()
```

```python
#after graph assessment drop irrelevant columns with minimal effect
#df = df.drop(["wind","visibility","functional"], axis=1)
#df["hour"] = np.array([np.sin(2*np.pi*hour/24) for hour in df["hour"]])

X = df.drop(["bike_count"], axis=1).values
y = np.log1p(df["bike_count"].values)
```

```python
# train, valid, test = np.split(df.sample(frac=1), [int(0.6*len(df)), int(0.8*len(df))])
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from xgboost import XGBRegressor
```

```python
xgb = XGBRegressor(
    n_estimators=500,
    learning_rate=0.05,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)
model = xgb
```

```python
from lightgbm import LGBMRegressor

lgbm = LGBMRegressor(
    n_estimators=500,
    learning_rate=0.05,
    random_state=42
)
model = lgbm
```

```python
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
print(r2e(y_test, y_pred))

print(model.score(x_train, y_train))
```
